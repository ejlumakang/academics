# Task 1: Global and Local Threshold

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Load the image in grayscale
    image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

    # Global Thresholding (Otsu's Method)
    _, global_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Local (Adaptive) Thresholding
    local_thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                         cv2.THRESH_BINARY, 11, 2)

    # Display results
    titles = ['Original Image', 'Global Thresholding', 'Local (Adaptive) Thresholding']
    images = [image, global_thresh, local_thresh]

    plt.figure(figsize=(10, 5))
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.show()

if __name__ == "__main__":
    main()
```

# Task 2: Edge Detection

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow

def edge_segmentation(image_path):
    # Load the image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Canny Edge Detection
    edges = cv2.Canny(gray, 100, 200)

    # Display the original and edge-detected images
    cv2_imshow(image)
    cv2_imshow(edges)


    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
image_path = 'image.jpg'  # Replace with your image path
edge_segmentation(image_path)
```

# Task 3: Region-based Segmentation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def region_based_segmentation(image_path):
    image = cv2.imread(image_path, 0)
    _, binary = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    # Find connected components
    num_labels, labels = cv2.connectedComponents(binary)

    plt.imshow(labels, cmap='jet')
    plt.title(f'Segmented Regions ({num_labels} regions)')
    plt.axis('off')
    plt.show()

# Example usage
image_path = 'image.jpg'  # Change this to your image path
region_based_segmentation(image_path)
```

2. Global thresholding uses a single threshold value for the entire image and works best on images with even lighting and clear contrast between the foreground and background. This technique can be used in **Optical Character Recognition (OCR)** for separating text from the background in scanned documents and **Object Detection in Industrial Inspection** for identifying product defects by highlighting contrasting areas.

  While Local thresholding uses different thresholds for each pixel based on nearby pixels, making it effective for images with uneven lighting or textures. For example, in **Medical Image Analysis**, it can enhance the contrast of specific features within images. In **License Plate Recognition**, it helps segment characters from the background despite uneven illumination or dirt.



3. I think that each technique excel in different contexts. Edge-based methods excel with clear object boundaries, ideal for object recognition. Region-based methods are better suited for complex, low-contrast images where boundaries are less defined, like in medical or satellite imagery. Ultimately, the best choice depends on the specific problem, image characteristics, and computational resources.
