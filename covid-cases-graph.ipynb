```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/machine learning/covid cases.xlsx')

df
```

```python
# Initialize elements
x = df['Location']
y1 = df['Confirmed_Cases']
y2 = df['Recoveries']
y3 = df['Deaths']
labels = ['Confirmed Cases', 'Recoveries', 'Deaths']
colors = ['blue', 'red', 'green']
```

```python
# Scatterplot

# Create scatterplot
plt.scatter(x, y1, color=colors[0], marker='.', label=labels[0])
plt.scatter(x, y2, color=colors[1], marker='.', label=labels[1])
plt.scatter(x, y3, color=colors[2], marker='.', label=labels[2])

# Labels, title, and legend
plt.title('COVID-19 Data in Cavite')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Count')
plt.legend()

# Show plot
plt.show()
```

```python
# Line Graph

# Create line graph
plt.plot(x, y1, marker='.', label=labels[0])
plt.plot(x, y2, marker='.', color=colors[1], label=labels[1])
plt.plot(x, y3, marker='.', color=colors[2], label=labels[2])

# Labels, title, and legend
plt.title('COVID-19 Data in Cavite')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Count')
plt.legend()

# Show plot
plt.show()
```

```python
# Bar Graph

# Get number of location
n = len(df['Location'])

bar_width = 0.25

# Positions on the x-axis where the bars will be placed
index = np.arange(n)

# Create bar graph
plt.bar(index - bar_width, y1, width=bar_width, label=labels[0])  # Left bar
plt.bar(index, y2, width=bar_width, color=colors[1], label=labels[1]) # Center bar
plt.bar(index + bar_width, y3, width=bar_width, color=colors[2], label=labels[2]) # Right bar

# Labels, title, and legend
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('COVID-19 Data in Cavite')
plt.xticks(index, df['Location'], rotation=90)
plt.legend()

# Show plot
plt.show()
```

```python
# Pie Chart

# Create pie chart
sizes = [y1.sum(), y2.sum(), y3.sum()]
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Labels, title, and legend
plt.title('Total COVID-19 Data Proportions')

# Show plot
plt.show()
```

```python
# Histogram (Stacked version)

# Create stacked histogram
plt.hist([y1, y2, y3], bins=10, stacked=True, color=colors, edgecolor='black', label=labels)

# Labels, title, and legend
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.title('COVID-19 Data in Cavite')
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
